## Modelos: vamos a trabajar con cuatro modelos de predicción a la vez y así puedo ver cuando mejoro resultados

### Modelo GaussianNB con un train_test_split incorporado y bucle que genera 100 distribuciones aleatorias y entrega la media del resultado:

### PD: en la entrega de la semana que viene usaré el método kfold y se tendrá en cuenta la varianza

```python
def testeoDataSetConGaussianNB(X,y):

    from sklearn.model_selection import train_test_split
    from sklearn.naive_bayes import GaussianNB
    errorNB = 0
    count = 0
    for random in range(0,99,1):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random)
        gnb = GaussianNB()
        y_pred = gnb.fit(X_train, y_train).predict(X_test)
#         print("Number of mislabeled points out of a total %d points : %d = %d percent error ratio" % (X_test.shape[0], (y_test != y_pred).sum(),100*(y_test != y_pred).sum()/X_test.shape[0]))
        errorNB += (100*(y_test != y_pred).sum()/X_test.shape[0])
        count += 1
    print("media de error en %d distribuciones distintas %d por ciento" %(count, errorNB/count))
    count = 0
    
```

```python
testeoDataSetConGaussianNB(X_df_v1,y_df)
```

### Modelo KNeighborsClassifier con un train_test_split incorporado y bucle que genera 100 distribuciones aleatorias y entrega la media de los resultados en forma de error:

```python
def testeoDataSetConKNeighborsClassifier(X,y):
    from sklearn.model_selection import train_test_split
    from sklearn.neighbors import KNeighborsClassifier
    errorKNN = 0
    count = 0
    for random in range(0,99,1):
        X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=random)
        neigh = KNeighborsClassifier(n_neighbors=5)
        y_pred = neigh.fit(X_train, y_train).predict(X_test)
    #     print("Number of mislabeled points out of a total %d points : %d = %d percent error ratio" % (X_test.shape[0], (y_test != y_pred).sum(),100*(y_test != y_pred).sum()/X_test.shape[0]))
        errorKNN += (100*(y_test != y_pred).sum()/X_test.shape[0])
        count += 1
    print("media de error en %d distribuciones distintas de %d por ciento" %(count, errorKNN/count))
    count = 0
```

```python
testeoDataSetConKNeighborsClassifier(X_df_v1,y_df)
```

### Modelo LogisticRegression con un train_test_split incorporado y bucle que genera 100 distribuciones aleatorias y entrega la media del resultado:

```python
def testeoDataSetConLogisticRegression(X,y):
    from sklearn.model_selection import train_test_split
    from sklearn.linear_model import LogisticRegression
    errorLR = 0
    count = 0
    for random in range(0,99,1):
        X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=random)
        LogReg = LogisticRegression(solver="liblinear",max_iter = 5000, random_state=0, multi_class="ovr")
        y_pred = LogReg.fit(X_train, y_train).predict(X_test)
#         print("Number of mislabeled points out of a total %d points : %d = %d percent error ratio" % (X_test.shape[0], (y_test != y_pred).sum(),100*(y_test != y_pred).sum()/X_test.shape[0]))
        errorLR += (100*(y_test != y_pred).sum()/X_test.shape[0])
        count += 1
    print("media de error en %d distribuciones distintas de %d por ciento" %(count, errorLR/count))
    count = 0
```

```python
testeoDataSetConLogisticRegression(X_df_v1,y_df)
```

### Modelo DecisionTreeClassifier con un train_test_split incorporado y bucle que genera 100 distribuciones aleatorias y entrega la media del resultado:

```python
def testeoDataSetConDecisionTreeClassifier(X,y):
    from sklearn.model_selection import train_test_split
    from sklearn import tree
    errorTree = 0
    count = 0
    for random in range(0,99,1):
        X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=random)
        DecTree = tree.DecisionTreeClassifier()
        y_pred = DecTree.fit(X_train, y_train).predict(X_test)
#         print("Number of mislabeled points out of a total %d points : %d = %d percent error ratio" % (X_test.shape[0], (y_test != y_pred).sum(),100*(y_test != y_pred).sum()/X_test.shape[0]))
        errorTree += (100*(y_test != y_pred).sum()/X_test.shape[0])
        count += 1
    print("media de error en %d distribuciones distintas de %d por ciento" %(count, errorTree/count))
    count = 0
    
```

```python
testeoDataSetConDecisionTreeClassifier(X_df_v1,y_df)
```

### Método de introducción de un dataset a los 4 modelos con el fín de ver que modelo funciona mejor:

```python
def TesteoModelos(X, y):
    print("modelo naive bayes gaussiano")
    testeoDataSetConGaussianNB(X,y)
    print("modelo arbol de decisiones de clasificación")
    testeoDataSetConDecisionTreeClassifier(X,y)
    print("modelo arbol de regresión logistica")
    testeoDataSetConLogisticRegression(X,y)
    print("modelo de K vecinos")
    testeoDataSetConKNeighborsClassifier(X,y)
```

```python
TesteoModelos(X_df_v1,y_df)
```
